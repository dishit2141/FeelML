```python
# download the english words
from os.path import basename, exists

def download(url):
    filename = basename(url)
    if not exists(filename):
        from urllib.request import urlretrieve
        local, _ = urlretrieve(url, filename)
        print('Downloaded ' + local)

download('https://github.com/AllenDowney/DSIRP/raw/main/american-english')
```

```python
#read words from list
def read_words(filename):
  res = set()
  for line in open(filename):
    for word in line.split():
      res.add(word.strip().lower())
  return res

words_list = read_words('/content/american-english')
len(words_list)
```

```python
"""
Logic for anagrams:
1)get list of words
2)loop through the list and check with the actuall logic for each words
3)store and return all possible options
"""
```

```python
#check anagrams between 2 words
def is_anagrams(word1,word2):
  if(len(word1) == len(word2)):
      if(sorted(word1) == sorted(word2)):
        return word2
      else:
        return False
  else:
    return False

#now will be checking anagrams on list
def is_anagrams_from_list(word,words_list):
  res = []
  for w in words_list:
    if is_anagrams(word,w):
      res.append(w)
  return res
```

```python
"""
Exercise: Find the longest word with at least one anagram. Suggestion: use the key argument of sort or sorted (see here).
"""
def longest_word_with_anagrams(words_list):
    longest_word = ''

    for word in words_list:
        # Check if the current word is longer than the longest found so far
        if len(word) > len(longest_word):
            # Check if it has at least one anagram
            if is_anagrams_from_list(word, words_list):
                longest_word = word

    if longest_word:
        print(f"The longest word with at least one anagram is: {longest_word}")
    else:
        print("No word with anagrams found.")

longest_word_with_anagrams(words_list)
```

```python
"""
Exercise: Find the largest list of words that are anagrams of each other.
"""
def find_largest_list_of_anagrams(words_list):
    anagram_dict = {}

    # Group words by their sorted letters
    for word in words_list:
        sorted_word = ''.join(sorted(word))
        if sorted_word in anagram_dict:
            anagram_dict[sorted_word].append(word)
        else:
            anagram_dict[sorted_word] = [word]

    # Find the largest list of anagrams
    largest_anagram_group = []
    for group in anagram_dict.values():
        if len(group) > len(largest_anagram_group):
            largest_anagram_group = group

    return largest_anagram_group

find_largest_list_of_anagrams(words_list)
```
